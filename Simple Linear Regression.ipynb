### Performing Linear Regression

```python
df
```

```python
x=df.income
```

```python
y=df.happiness
```

```python
import numpy as np
```

```python
x=np.array(x).reshape(-1,1)
x
```

```python
y=np.array(y).reshape(-1,1)
y
```

```python
# import LinearRegression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LinearRegression()
```

```python
# Fit the model to the data
lr.fit(x,y)
```

```python
import statsmodels.api as sm
```

```python
# to get intercept of the model
c=lr.intercept_
c
```

```python
## to get gradient of the model
m=lr.coef_
m
```

```python
# to get rsquared of the model
lr.summary()
```

```python
# before fitting linear regression model it is common that you spliting the data in training and testung but in this example
# i will be skip that step 
```

### predictions

```python
y_pre=lr.predict(x)
```
